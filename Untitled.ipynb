```python
import numpy as np
import torch
import matplotlib.pyplot as plt


ckpt_path = './lightning_logs/version_1/checkpoints/epoch=81-step=33395.ckpt'
ckpt = torch.load(ckpt_path)
```

```python
from stamp_ae import StampAE, IndexDataset

model = StampAE()

model.load_state_dict(ckpt['state_dict'])
model.cuda()
model.eval()

ds = IndexDataset()

for i in range(30):
    x = torch.from_numpy(ds[i])
    y = model.decoder(x[None,:].cuda())[0]
    y = y.detach().cpu().numpy()
    y /= y.max()
    y = np.transpose(y,[1,2,0])
    
    z = np.median(y)
    print(z)
    img = (y > z).astype(np.float32)
    plt.imshow(img)
    plt.show()
    plt.close()
```
